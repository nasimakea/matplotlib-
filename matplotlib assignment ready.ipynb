# Here are the examples

```python
#Line plot
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(2,20,10)
y=np.random.randint(2,20,10)
```

```python


plt.plot(x,y,color='r')
plt.xlabel("x random data")
plt.ylabel("y random data")
plt.title("random data vs random data")
```

```python
#scatter plot
```

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
c=['r','b','g','y','b','g','g']

plt.scatter(day,no,color=c)
plt.title("scatter plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No",fontsize=15)
plt.show()
```

```python
x=np.random.randint(0,20,10)
```

```python
plt.hist(x)
plt.xlabel("x_random_data")
plt.ylabel("y_random_data")
plt.title("x data vs y data")
plt.show()
```

```python
#pie chart
```

```python
y=['Bihar','Uttar pradesh','Delhi','Mumbai']
x=[85,70,30,20]
c=['r','b','g','y']



plt.pie(x,labels=y,colors=c,radius=1.5)


plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])

# ANS:-

```python

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

# A bar plot (or bar chart) is a type of data visualization that presents categorical data with rectangular bars. The lengths of the bars are proportional to the values they represent. Bar plots are commonly used to compare the values of different categories or to show the distribution of a single categorical variable.

```python
plt.title("Company vs Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
c=['g','r','y','b']
plt.bar(company,profit,color=c)
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)

# ANS:-

# The box plot is used to visulalize data and helps to detects the outliers , it has lower bound , upper bound , first quartile , median and third quartile, but mainly focoused on detedting outliers visually

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.subplot(2,2,1)
plt.boxplot(box1)

plt.subplot(2,2,2)
plt.boxplot(box2)

plt.show()
```
